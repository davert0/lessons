Полный стакан с водой весит 300 граммов, а вес полупустого соответствует весу двух пустых стаканов. Сколько весит пустой стакан?

````python
half_water = Real("half_water")
empty = Real("empty")

solve(
    empty + 2*half_water == 300,
    empty + half_water == empty + empty
)

````

В квартире живут домашние животные: собаки и кошки. Из всех животных только одно не является собакой. При этом все питомцы, кроме одного, — кошки. Сколько всего кошек и собак?

````python
dogs = Int("dogs")
cats = Int("cats")
total = Int("total")

solve(
    dogs + 1 == total,
    total - 1 == cats,
    dogs != 0,
    cats != 0
)
````

Маме и дочке вместе 28 лет. Мама старше дочери на 22 года. Сколько лет дочке и сколько маме?

````python
mother = Int("Mother")
daughter = Int("Daughter")

solve(
    mother + daughter == 28,
    mother - daughter == 22
)
````

На шахматной доске надо расставить восемь ферзей так, чтобы никто друг друга не атаковал.

Один из вариантов расстановки — поставить по одному ферзю в каждой строке на следующие клетки:

**a1, e2, h3, f4, c5, g6, b7, d8**

```text
8 | . . . ♛ . . . .
7 | . ♛ . . . . . .
6 | . . . . . . ♛ .
5 | . . ♛ . . . . .
4 | . . . . . ♛ . .
3 | . . . . . . . ♛
2 | . . . . ♛ . . .
1 | ♛ . . . . . . .
    a b c d e f g h
```

В виде номеров столбцов для каждой строки:

```text
[1, 5, 8, 6, 3, 7, 2, 4]
```

Ферзи не атакуют друг друга, потому что:

* в каждой строке находится ровно один ферзь;
* в каждом столбце находится ровно один ферзь;
* ни у одной пары ферзей не совпадают диагонали.

Для двух ферзей в клетках `(строка₁, столбец₁)` и `(строка₂, столбец₂)` совпадение диагонали проверяется условием:

```text
|строка₁ − строка₂| = |столбец₁ − столбец₂|
```

В этой расстановке такого равенства ни для одной пары нет.
